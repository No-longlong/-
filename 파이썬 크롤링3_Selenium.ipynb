```python
# 무한 스크롤
import time

SCROLL_PAUSE_TIME = 10

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight") 

while True:
    # Scroll down to bottom                                                   
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)                                            
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")  
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height         
    new_height = browser.execute_script("return document.body.scrollHeight")

    if new_height == last_height:                                             
        break

    last_height = new_height
```

```python
# 상명대는 소통하라
def sm_communication(url):

    while True:

        browser.get('https://www.naver.com/')

        time.sleep(3)

        browser.get(url)

        scroll_time = 15

        last_height = browser.execute_script("return document.body.scrollHeight") 

        while True:
            time.sleep(scroll_time)
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            time.sleep(scroll_time)                                            
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")

            bro_height = browser.execute_script("return document.body.scrollHeight")

            if bro_height == last_height:                                             
                break

            last_height = bro_height

sm_communication('https://search.naver.com/search.naver?where=nexearch&sm=tab_lve.ag20sgrpmamsipen0spm&ie=utf8&query=%EC%83%81%EB%AA%85%EB%8C%80%EB%8A%94+%EC%86%8C%ED%86%B5%ED%95%98%EB%9D%BC')
```

### 함수화

```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
#브라우저 띄우는 함수 

def _browser(headless=False):
    chrome_options = Options()
    if headless is True:
        chrome_options.add_argument("--headless")
    browser = webdriver.Chrome('./webdriver/chrome/chromedriver.exe', options=chrome_options)
    browser.implicitly_wait(3)
    browser.set_window_size(1920, 1280)
    return browser
    
def _get_page(url):
    browser.get(url)
    
def login_evertime(browser, _id,_pw):
    ele = browser.find_element_by_css_selector('input[name="userid"]')
    p_ele = browser.find_element_by_css_selector('input[name="password"]')
    ele.send_keys(_id)
    p_ele.send_keys(_pw)
    submit_ele = browser.find_element_by_css_selector('#container > form > p.submit > input')
    submit_ele.click()
    
def _click_hot_post(browser, first_or_second):
    """
    실시간 인기글은 2가지인데 둘중하나 선택해서 이동 후 html return
    """
    if first_or_second == 'first':
        hot_ele = browser.find_element_by_css_selector('#container > div.rightside > div:nth-child(2) > div > a:nth-child(2)')
    elif first_or_second =='second':
        hot_ele = browser.find_element_by_css_selector('#container > div.rightside > div:nth-child(2) > div > a:nth-child(3)')
    hot_ele.click()
    time.sleep(2)
    return browser.page_source

def data_parser(page_source):
    soup = BeautifulSoup(page_source, "html.parser")
    post = soup.select_one('#container > .wrap.articles > article > a.article')
    title = post.h2.text
    contents=""
    for p in post.select('p'):
        contents += p.text
    vote = post.select_one('ul:nth-child(7) > li.vote').text
    comment = post.select_one('ul:nth-child(7) > li.comment').text
    scrap = post.select_one('ul:nth-child(7) > li.scrap').text
    
    data_dict = {
        "제목":title,
        "내용":contents,
        "공감":vote,
        "댓글수":comment,
        "스크랩수": scrap
                 }
    return data_dict
```

```python
def main():
    browser = _browser()
    browser.get('https://everytime.kr/login')
    login_evertime(browser, 'ID', 'PW')
    page_source=_click_hot_post(browser, 'second')
    data_dict = data_parser(page_source)
    print(data_dict)
```

```python
main()
```
