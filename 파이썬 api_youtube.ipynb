```python
# video 서비스 사용

url = 'https://www.googleapis.com/youtube/v3/videos'
params = {
    'part': 'id,snippet,statistics',
    'id': 'NCeazanb_jE', # id는 따옴표 사이에 여러개의 id값을 가질 수 있음.
    'key': api_key
}
```

```python
# API 요청
res = requests.get(url, headers=headers, params=params)
```

```python
video_res = res.json()
```

> - json()으로부터 비디오 ID, 제목, 업로드 날짜, 수집한 날짜, 동영상 설명, 채널, 좋아요, 싫어요, 조회수를 정리할 필요성이 있다.

```python
# 모든 resource는 'items'에 있다는 점.
video_res['items'][0]['snippet'] # 기본 정보
video_res['items'][0]['statistics'] # 통계 정보
```

```python
# pandas에 저장하기 위해 dict로 만들기
snippet = video_res['items'][0]['snippet']

statistics = video_res['items'][0]['statistics']


save_params = {
    'title': snippet['title'],
    'description': snippet['description'],
    'publishedAt': snippet['publishedAt'],
    'channelId': snippet['channelId'],
    'viewCount': statistics['viewCount'],
    'likeCount': statistics['likeCount'],
    'disilkeCount': statistics['dislikeCount'],
    'commentCount': statistics['commentCount']
}

print(save_params)
```

```python
# 추천하지는 않는 방법: 비디오 목록 50개 중 하나를 위해서 요청하는 것은 효율적이지 못함.

df = pd.DataFrame()
for video_id in video_list:
    url = 'https://www.googleapis.com/youtube/v3/videos'
    params = {
        'part': 'id,snippet,statistics',
        'id': 'NCeazanb_jE',
        'key': api_key
    }
    
    # API 요청
    res = requests.get(url, headers=headers, params=params)
    video_res = res.json()
    video_id = video_res['items'][0]['id']
    snippet = video_res['items'][0]['snippet']

    statistics = video_res['items'][0]['statistics']

    
    save_params = {
        'id': video_id,
        'title': snippet['title'],
        'description': snippet['description'],
        'publishedAt': snippet['publishedAt'],
        'channelId': snippet['channelId'],
        'viewCount': statistics['viewCount'],
        'likeCount': statistics['likeCount'],
        'disilkeCount': statistics['dislikeCount'],
        'commentCount': statistics['commentCount']
    }
    df = df.append(save_params, ignore_index=True)
```

```python
# search API

url = 'https://www.googleapis.com/youtube/v3/search'
params = {
    'part': 'id, snippet',
    'q': '금융',
    'maxResults': 50,
    'key': api_key
}
```

```python
res = requests.get(url, headers=headers, params=params)
```

```python
search_result = res.json()
```

```python
search_list = search_result['items']
```

```python
# 제목, 발행날짜(업로드 날짜), 설명, 썸네일(URL), 채널 정보

search_df = pd.DataFrame()

for search_obj in search_list:
    snippet = search_obj['snippet']
    save_params = {
        'title': snippet['title'],
        'description': snippet['description'],
        'publishedAt': snippet['publishedAt'],
        'thumbnail': snippet['thumbnails']['default']['url'],
        'channelTitle': snippet['channelTitle']
    }
    
    search_df = search_df.append(save_params, ignore_index=True)
```
